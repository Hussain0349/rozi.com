```python

import numpy as np
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from bs4 import BeautifulSoup
import time
s = Service(r'C:\Users\Qur\Desktop\chromedriver.exe')
driver = webdriver.Chrome(service=s)
link = driver.get('https://www.rozee.pk/job/jsearch/q/all/fin/1/fpn/0')

# List for Temperaroy store
job_title = []
location = []
salary_date_status = []
skills_required = []
while True:
    html = driver.page_source
    soup = BeautifulSoup(html,'lxml')
    divs = soup.find_all('div',class_ = 'job')
    for i in divs:
        try:
            job_title.append(i.find('h3',class_ ='s-18').find('bdi').text)
            location.append(i.find('bdi',class_="float-left").text)
            salary_date_status.append(i.find('div',class_ = 'col-md-12 float-left').text)
            skills_required.append(i.find('div',class_ = 'job-dtl clearfix').text)
        except:
            print(np.nan)
    try:
        
        next_button_xpath = '//a[@class="next" and contains(text(), "Next")]'
        driver.find_element(by = By.XPATH,value=next_button_xpath).click()
        # next_button = WebDriverWait(driver, 10).until(
        # EC.element_to_be_clickable((By.XPATH, next_button_xpath))
        # )
        # next_button.click()
        time.sleep(3)
    except:
        print('There is issue while clicking on Button Next')
        break

```

```python
import pandas as pd 
df = pd.DataFrame({
    'Job Title': job_title,
    'salary_date_status': salary_date_status,
    'location': location,
    'skills_required': skills_required
})
```

```python
df.to_csv('Job_data.csv')
```
